```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Method to read randomwalk file

```python
def read_randomwalk_file(file_path):
    with open(file_path, 'r') as file:
        walks = []
        for line in file:
            yield line.strip().split('|')
```

Reading Events Randomwalks

```python
walks_generator = read_randomwalk_file('C:\\Users\\orobatic\\Desktop\\books\\Tamer\\Social_Profile_Analysis\\data\\storage\\raw\\Last.fm Multigraph\\events\\randomwalk-0_anonymized')
walks_generator1 = read_randomwalk_file('C:\\Users\\orobatic\\Desktop\\books\\Tamer\\Social_Profile_Analysis\\data\\storage\\raw\\Last.fm Multigraph\\events\\randomwalk-1_anonymized')
walks_generator2 = read_randomwalk_file('C:\\Users\\orobatic\\Desktop\\books\\Tamer\\Social_Profile_Analysis\\data\\storage\\raw\\Last.fm Multigraph\\events\\randomwalk-2_anonymized')
walks_generator3 = read_randomwalk_file('C:\\Users\\orobatic\\Desktop\\books\\Tamer\\Social_Profile_Analysis\\data\\storage\\raw\\Last.fm Multigraph\\events\\randomwalk-3_anonymized')

walks = []
for walkg0,walkg1, walkg2, walkg3 in zip(walks_generator, walks_generator1, walks_generator2, walks_generator3):
    walks.append(walkg0)
    walks.append(walkg1)
    walks.append(walkg2)
    walks.append(walkg3)
print(f"Total walks read: {len(walks)}")
print("Completed reading randomwalk file.")
```

```python
random_walks_df = pd.DataFrame(walks, columns=['Source', 'type', 'Target'])
print(random_walks_df.head())
print("DataFrame created from random walks.")
```

```python
sources = random_walks_df['Source'].unique()
targets = random_walks_df['Target'].unique()

print(f"Unique sources: {len(sources)}")
print(f"Unique targets: {len(targets)}")

for source in sources:
    source_connection_count = random_walks_df[random_walks_df['Source'] == source].value_counts().sum()
    print(f"Source: {source}, Number of connections: {source_connection_count}")
```

```python
def make_keywords_list(keywords_file_path):
    with open(keywords_file_path, 'r') as file:
        for line in file:
            words = line.strip().split()
            for word in words:
                yield word
```
